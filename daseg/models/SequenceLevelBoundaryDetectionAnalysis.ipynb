```python
import os, sys
import numpy as np
import pickle as pkl
import sklearn.metrics.pairwise as sklearn_pairwise
import matplotlib.pyplot as plt
import time
from nltk.metrics.segmentation import pk
from scipy.special import softmax
```

```python
results_pkl = '/dih4/dih4_2/jhu/Raghu/topic_seg_expts/TrueCasing_CV_1_LabelScheme_Segmentation_text_model_topicseg_Morethan2TasksArch_SeqClassif_BERT_loss_wts_1_0_Epochs_25_bert-base-polish-cased-v1_logisfera_telco_ForI-Scheme_WithSEP_OnlySmoothSegmentation_/topicseg_Morethan2TasksArch_SeqClassif_BERT_TrueCasing_42/results.pkl'
results = pkl.load(open(results_pkl, 'rb'))

for doc_ind, utt_true_pred_labels in enumerate(zip(results['true_labels_op0'], results['predictions_op0'])):
#     print(doc_ind, utt_true_pred_labels)
    utt_true_labels = utt_true_pred_labels[0]
    utt_pred_labels = utt_true_pred_labels[1]
    utt_true_labels = [1 if i == 'Boundary' else 0 for i in utt_true_labels]
    utt_pred_labels = [1 if i == 'Boundary' else 0 for i in utt_pred_labels]
    utt_model_probs = softmax(results['logits_op0'][doc_ind], axis=-1)
    utt_model_pred_boundary_probs = utt_model_probs[:len(utt_true_labels)][:, 0]
#     print(utt_model_pred_boundary_probs)
    plt.figure()
    plt.plot(utt_true_labels, 'g')
    plt.plot(utt_pred_labels, 'r-')
    plt.plot(utt_model_pred_boundary_probs, 'r*--')
    plt.plot([0.5]*len(utt_true_labels), 'b')
    plt.ylim(-0.2, 1.2)
    
    plt.legend(['Ground truth', 'predicted', 'confidence'])
    plt.savefig('temp_fig')
    plt.show()
    time.sleep(1)
```
